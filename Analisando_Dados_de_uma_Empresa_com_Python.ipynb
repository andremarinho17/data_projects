Uma das grandes aliadas da tomada de decisão baseada em dados é a programação. Sim, isso mesmo, a programação! Com a programação, podemos simplificar nosso processo de análise de dados e obter insights poderosos, principalmente utilizando a linguagem Python.

Python é uma <a href="https://rockcontent.com/br/blog/linguagem-de-programacao/"> linguagem de programação</a> <a href="https://blog.geekhunter.com.br/quais-sao-os-paradigmas-de-programacao/">multiparadigma</a> e que nos fornece uma variedade de ferramentas através de suas bibliotecas e dentre elas, está a análise de dados. Uma biblioteca do Python excelente para se trabalharmos é o Pandas, que é extremamente poderosa e útil para análise de dados.

Para ilustrar os dois benefícios que citei, irei analisar os dados de uma empresa fictícia usando Python e utilizarei a seguinte planilha, que contem três abas: Estoque, Vendas e Funcionários

Primeiramente, vamos importar o Pandas

```
import pandas as pd
```

 Agora, vamos importar a planilha do excel

```
planilha = pd.read_excel("/content/controle_da_empresa.xlsx", None)
```

Por fim, basta executar a planilha


```
planilha
```

A seguir, as três abas da planilha

**Aba Estoque**


```
planilha["Estoque"]
```

**Aba Vendas**

```
planilha["Vendas"]
```

**Aba Funcionários**

```
planilha["Funcionários"]
```

Com Python, posso acessar uma coluna específica de minha planilha e efetuar diversas operações de analise de dados.

```
planilha["Vendas"]["Total de Vendas"]
```

Somando os valores de uma coluna:

```
planilha["Vendas"]["Total de Vendas"].sum()
```

Calculando a média de um coluna:

```
planilha["Vendas"]["Total de Vendas"].mean()
```

Calculando a mediana de um coluna:

```
planilha["Vendas"]["Total de Vendas"].median()
```

Calculando o desvio padrão de um coluna

```
planilha["Vendas"]["Total de Vendas"].std()
```

Para simplificar as operações, posso passar uma aba da planilha para uma <a href="https://tableless.github.io/iniciantes/manual/js/variaveis.html#:~:text=Crie%2C%20manipule%20e%20use%20valores%20vari%C3%A1veis&text=Uma%20vari%C3%A1vel%20referencia%20a%20um,ser%C3%A3o%20usadas%20em%20seus%20programas.&text=Porque%20uma%20vari%C3%A1vel%20pode%20ter,a%20execu%C3%A7%C3%A3o%20de%20um%20programa."> variável </a>. 

Criando uma nova variável:

```
vendas = planilha["Vendas"]
```

Visualizando a variável criada:

```
vendas
```

Para obtermos uma melhor visualização do comportamento das vendas da empresa, podemos plotar um gráfico de linhas bem simples, com a Data da Venda no eixo X e o Total de Vendas no eixo Y.

```
vendas.plot("Data da Venda", "Total de Vendas")
```

Dessa forma, conseguimos observar como estão se comportando os dados em pequenos períodos de tempo. Por exemplo, entre os dias 13 e 21 de Janeiro, as vendas ficaram em baixa, apresentando alguns picos, porém inconstantes. Já a partir do dia 21, as vendas ganharam uma certa regularidade, se mantendo entre 15 e 20 mil reais.

Uma grande estratégia para melhorar a experiência do consumidor é estudar seu comportamento, como por exemplo, quanto ele gasta em média. Podemos obter esta informação através de um Histograma para visualizar a distribuição dos dados.

```
vendas.hist("Total de Vendas")
```

Observe que a distribuição se concentra em grande parte entre 0 e R$ 5.000, mas para obtermos um pouco mais de exatidão acerca do intervalo, podemos verificar qual é o valor mínimo do Total de Vendas da empresa.

```
# Verificando o valor mínimo do Total de Vendas 

vendas["Total de Vendas"].min()
```

Dessa forma, conseguimos chegar à conclusão que a maioria de nossos consumidores gasta entre 1.200 e 5 mil reais. A empresa pode formular novas estratégias sobre a precificação de seus produtos para assim, melhorar a experiência e satisfação de seus clientes.

Observe que as datas das vendas são do mês de janeiro de 2020. Neste caso, podemos fazer uma análise de dados mais precisa para prever o comportamento das vendas e tomar decisão. Para isso, podemos utilizar também o Histograma para visualizar a distribuição de dados semanal de vendas por dia.

```
vendas["Data da Venda"].apply(lambda x: x.weekday()).hist(bins=7)
```

O gráfico acima começa na segunda-feira (0) e vai até o domingo (6). Observe que a maior parte das pessoas realiza compras nas terças e quintas-feiras. Assim, a empresa poderia investir mais em propaganda e campanhas de marketing nestes dias, sabendo que a probabilidade de aumentar seu faturamento é altíssima. Ou seja, a empresa antevem as tendências, se prepara para elas e concorrentemente, aperfeiçoa a experiência e satisfação do consumidor.
